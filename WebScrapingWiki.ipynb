```
from bs4 import BeautifulSoup
from requests import get
import re
```

```
r = get("https://th.wikipedia.org/wiki/%E0%B8%82%E0%B9%89%E0%B8%AD%E0%B8%A1%E0%B8%B9%E0%B8%A5%E0%B8%82%E0%B8%99%E0%B8%B2%E0%B8%94%E0%B9%83%E0%B8%AB%E0%B8%8D%E0%B9%88")
r.text
```

```
content = BeautifulSoup(r.text).find_all('div', class_="mw-parser-output")
```

```
content[0].find_all('p')[0].text
```

```
text = re.sub(r"\n","",content[0].find_all('p')[0].text)
text = re.sub(r"\[[0-9]\]","",text)
```
